```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fungsi untuk membaca citra
def read_image(file_path):
    image = imageio.v3.imread('/content/rgb.jpeg')  # Membaca citra RGB
    return image

# Fungsi untuk mengenkripsi citra
def encrypt_image(image, key):
    # Pastikan key memiliki ukuran yang sesuai
    key = np.resize(key, image.shape)
    encrypted_image = np.bitwise_xor(image, key)  # XOR antara citra dan kunci
    return encrypted_image

# Fungsi untuk mendekripsi citra
def decrypt_image(encrypted_image, key):
    # Gunakan XOR yang sama untuk mendekripsi
    key = np.resize(key, encrypted_image.shape)
    decrypted_image = np.bitwise_xor(encrypted_image, key)
    return decrypted_image

# Main program
file_path = "input_image.jpg"  # Nama file yang Anda unggah

# Membaca citra asli
image = read_image(file_path)

# Menampilkan citra asli
plt.figure(figsize=(10, 10))
plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title("Citra Asli")
plt.axis('off')

# Membuat kunci enkripsi (array random dengan dimensi yang sesuai)
np.random.seed(42)  # Seed untuk menghasilkan angka acak yang sama
key = np.random.randint(0, 256, size=image.shape, dtype=np.uint8)

# Mengenkripsi citra
encrypted_image = encrypt_image(image, key)

# Menampilkan citra terenkripsi
plt.subplot(1, 3, 2)
plt.imshow(encrypted_image)
plt.title("Citra Terenkripsi")
plt.axis('off')

# Mendekripsi citra
decrypted_image = decrypt_image(encrypted_image, key)

# Menampilkan citra terdekripsi
plt.subplot(1, 3, 3)
plt.imshow(decrypted_image)
plt.title("Citra Terdekripsi")
plt.axis('off')

plt.show()
```

```python
# Menyimpan hasil
imageio.v3.imwrite("encrypted_image.png", encrypted_image)
imageio.v3.imwrite("decrypted_image.png", decrypted_image)

# Mengunduh hasil
from google.colab import files
files.download("encrypted_image.png")
files.download("decrypted_image.png")
```
